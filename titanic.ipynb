```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file = "Titanic-Dataset.csv"

df = pd.read_csv(file)
print(df.head(20).to_string(line_width=200))
```

```python
survived = df[df['Survived'] == 1]['Survived']
survived_c = survived.count()
print(f'Вижило: {survived_c} осіб')

not_survived = df[df['Survived'] == 0]['Survived']
not_survived_c = not_survived.count()
print(f'Не вижило: {not_survived_c} осіб')
```

```python
gender_counts = df['Sex'].value_counts()

# Виводимо кількість чоловіків та жінок
print("Кількість чоловіків:", gender_counts['male'])
print("Кількість жінок:", gender_counts['female'])

# Створюємо кругову діаграму для відображення співвідношення
plt.figure(figsize=(7, 7))
plt.pie(gender_counts, labels=['Чоловіки', 'Жінки'], autopct='%1.1f%%', startangle=140)
plt.title('Співвідношення чоловіків та жінок на Титаніку')
plt.legend()
plt.show()
```

```python
survived_male = df[(df['Survived'] == 1) & (df['Sex'] == 'male')].shape[0]
survived_female = df[(df['Survived'] == 1) & (df['Sex'] == 'female')].shape[0]

labels = ['Чоловіки', 'Жінки']
sizes = [survived_male, survived_female]
colors = ['lightblue', 'lightpink']

# Побудова кругової діаграми
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Порівняння кількості виживших чоловіків та жінок')
plt.axis('equal')
plt.legend()
plt.show()

survived_male = df[(df['Survived'] == 0) & (df['Sex'] == 'male')].shape[0]
survived_female = df[(df['Survived'] == 0) & (df['Sex'] == 'female')].shape[0]

# Створюємо дані для діаграми
labels = ['Чоловіки', 'Жінки']
sizes = [survived_male, survived_female]
colors = ['lightblue', 'lightpink']

# Побудова кругової діаграми
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Порівняння кількості загиблих чоловіків та жінок')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
male_survival_counts = df[(df['Sex'] == 'male') & (df['Survived'] == 1)].shape[0]
male_dead_counts = df[(df['Sex'] == 'male') & (df['Survived'] == 0)].shape[0]

female_survival_counts = df[(df['Sex'] == 'female') & (df['Survived'] == 1)].shape[0]
female_dead_counts = df[(df['Sex'] == 'female') & (df['Survived'] == 0)].shape[0]

# Створюємо стовпчасту діаграму для порівняння кількості виживших та загиблих чоловіків
plt.figure(figsize=(5, 5))
plt.bar(['Виживші', 'Загиблі'], [male_survival_counts, male_dead_counts], color=['blue', 'red'])
plt.title('Порівняння кількості виживших та загиблих чоловіків на Титаніку')
plt.xlabel('Статус')
plt.ylabel('Кількість')
plt.show()

plt.figure(figsize=(5, 5))
plt.bar(['Виживші', 'Загиблі'], [female_survival_counts, female_dead_counts], color=['pink', 'red'])
plt.title('Порівняння кількості виживших та загиблих жінок на Титаніку')
plt.xlabel('Статус')
plt.ylabel('Кількість')
plt.show()
```

```python
passengers_1_class = df[df['Pclass'] == 1]
passengers_1_class_count = passengers_1_class.shape[0]
print(f'Пасажири 1-го класу: {passengers_1_class_count}')

passengers_2_class = df[df['Pclass'] == 2]
passengers_2_class_count = passengers_2_class.shape[0]
print(f'Пасажири 2-го класу: {passengers_2_class_count}')

passengers_3_class = df[df['Pclass'] == 3]
passengers_3_class_count = passengers_3_class.shape[0]
print(f'Пасажири 3-го класу: {passengers_3_class_count}')
```

```python
survived_by_class_1 = df[(df['Survived'] == 1) & (df['Pclass'] == 1)]['Survived']
survived_by_class_1_c = survived_by_class_1.count()
print(f'Виживші з 1 класу: {survived_by_class_1_c}')

survived_by_class_1_v = survived_by_class_1_c/survived_c
print(f'Частота виживання людей з 1 класу: {survived_by_class_1_v.round(4)}')

survived_by_class_2 = df[(df['Survived'] == 1) & (df['Pclass'] == 2)]['Survived']
survived_by_class_2_c = survived_by_class_2.count()
print(f'\nВиживші з 2 класу: {survived_by_class_2_c}')

survived_by_class_2_v = survived_by_class_2_c/survived_c
print(f'Частота виживання людей з 2 класу: {survived_by_class_2_v.round(4)}')

survived_by_class_3 = df[(df['Survived'] == 1) & (df['Pclass'] == 3)]['Survived']
survived_by_class_3_c = survived_by_class_3.count()
print(f'\nВиживші з 3 класу: {survived_by_class_3_c}')

survived_by_class_3_v = survived_by_class_3_c/survived_c
print(f'Частота виживання людей з 3 класу: {survived_by_class_3_v.round(4)}')
```

```python
survived_young = df[(df['Survived'] == 1) & (df['Age'] < 30)]['Survived']
survived_young_c = survived_young.count()
print(f'Виживші до 30 років: {survived_young_c}')
not_survived_young = df[(df['Survived'] == 0) & (df['Age'] < 30)]['Survived']
not_survived_young_c = not_survived_young.count()
print(f'Загиблі до 30 років: {not_survived_young_c}')

survived_middle = df[(df['Survived'] == 1) & (df['Age'] >= 30) & (df['Age'] < 60)]['Survived']
survived_middle_c = survived_middle.count()
print(f'\nВиживші від 30 до 60 років: {survived_middle_c}')
not_survived_middle = df[(df['Survived'] == 0) & (df['Age'] >= 30) & (df['Age'] < 60)]['Survived']
not_survived_middle_c = not_survived_middle.count()
print(f'Загиблі від 30 до 60 років: {not_survived_middle_c}')

survived_old = df[(df['Survived'] == 1) & (df['Age'] >= 60)]['Survived']
survived_old_c = survived_old.count()
print(f'\nВиживші старші за 60 років: {survived_old_c}')
not_survived_old = df[(df['Survived'] == 0) & (df['Age'] >= 60)]['Survived']
not_survived_old_c = not_survived_old.count()
print(f'Загиблі старші за 60 років: {not_survived_old_c}')

age_categories = []
for age in df['Age']:
    if age < 30:
        age_categories.append('До 30')
    elif 30 <= age < 60:
        age_categories.append('Від 30 до 60')
    else:
        age_categories.append('Старше за 60')

# Рахуємо кількість пасажирів у кожній віковій категорії
age_counts = {}
for category in set(age_categories):
    age_counts[category] = age_categories.count(category)

# Створюємо дані для кругової діаграми
labels = age_counts.keys()
sizes = age_counts.values()
colors = ['lightblue', 'lightgreen', 'lightcoral']

# Побудова кругової діаграми
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Розподіл пасажирів по віковим категоріям')
plt.legend()
plt.axis('equal')
plt.show()
```

```python
survived_ages = df[df['Survived'] == 1]['Age']

survived_age_categories = []
for age in survived_ages:
    if age < 30:
        survived_age_categories.append('До 30')
    elif 30 <= age < 60:
        survived_age_categories.append('Від 30 до 60')
    else:
        survived_age_categories.append('Старше за 60')

survived_age_counts = {}
for category in set(survived_age_categories):
    survived_age_counts[category] = survived_age_categories.count(category)

# Створюємо дані для кругової діаграми
labels = survived_age_counts.keys()
sizes = survived_age_counts.values()
colors = ['lightblue', 'lightgreen', 'lightcoral']

# Побудова кругової діаграми
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Розподіл виживжих пасажирів по віковим категоріям')
plt.legend()
plt.axis('equal')
plt.show()



not_survived_ages = df[df['Survived'] == 0]['Age']

not_survived_age_categories = []
for age in not_survived_ages:
    if age < 30:
        not_survived_age_categories.append('До 30')
    elif 30 <= age < 60:
        not_survived_age_categories.append('Від 30 до 60')
    else:
        not_survived_age_categories.append('Старше за 60')

not_survived_age_counts = {}
for category in set(not_survived_age_categories):
    not_survived_age_counts[category] = not_survived_age_categories.count(category)

# Створюємо дані для кругової діаграми
labels = not_survived_age_counts.keys()
sizes = not_survived_age_counts.values()
colors = ['lightblue', 'lightgreen', 'lightcoral']

# Побудова кругової діаграми
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Розподіл загиблих пасажирів по віковим категоріям')
plt.legend()
plt.axis('equal')
plt.show()
```

```python
survived_by_S = df[(df['Survived'] == 1) & (df['Embarked'] == 'S')]['Survived']
survived_by_S_c = survived_by_S.count()
print(f'Виживші з порту Southampton: {survived_by_S_c}')

survived_by_C = df[(df['Survived'] == 1) & (df['Embarked'] == 'C')]['Survived']
survived_by_C_c = survived_by_C.count()
print(f'Виживші з порту Cherbourg: {survived_by_C_c}')

survived_by_Q = df[(df['Survived'] == 1) & (df['Embarked'] == 'Q')]['Survived']
survived_by_Q_c = survived_by_Q.count()
print(f'Виживші з порту Queenstown: {survived_by_Q_c}')
```

```python
survival_count_by_embarked = df.groupby(['Embarked', 'Survived']).size().unstack()
survival_count_by_embarked['Survival_Ratio'] = survival_count_by_embarked[1] / survival_count_by_embarked.sum(axis=1)

print("Кількість виживших та загиблих пасажирів, а також відношення кількості виживших до загиблих за портами посадки:")
print(survival_count_by_embarked)

# Графік для порівняння кількості виживших
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
survival_count_by_embarked[1].plot(kind='bar', color='green')
plt.title('Кількість виживших пасажирів за портами посадки')
plt.xlabel('Порт посадки')
plt.ylabel('Кількість виживших')

# Графік для порівняння кількості загиблих
plt.subplot(1, 2, 2)
survival_count_by_embarked[0].plot(kind='bar', color='red')
plt.title('Кількість загиблих пасажирів за портами посадки')
plt.xlabel('Порт посадки')
plt.ylabel('Кількість загиблих')

plt.tight_layout()
plt.show()
```

```python
df['Fare'].fillna(0, inplace=True)

plt.figure(figsize=(14, 7))

# Графік для виживших пасажирів
plt.subplot(1, 2, 1)
sns.histplot(df[df['Survived'] == 1]['Fare'], bins=50, kde=True, color='green')
plt.title('Розподіл вартості квитків серед виживших пасажирів')
plt.xlabel('Вартість квитка')
plt.ylabel('Кількість пасажирів')

# Графік для загиблих пасажирів
plt.subplot(1, 2, 2)
sns.histplot(df[df['Survived'] == 0]['Fare'], bins=50, kde=True, color='red')
plt.title('Розподіл вартості квитків серед загиблих пасажирів')
plt.xlabel('Вартість квитка')
plt.ylabel('Кількість пасажирів')

plt.tight_layout()
plt.show()
```

```python
mean_fare_by_class = df.groupby('Pclass')['Fare'].mean()

print("Середня вартість квитка для кожного класу:")
print(mean_fare_by_class.round(4))

sum_for_1_class = df[df['Pclass'] == 1]['Fare'].sum()
print(f'\nСума цін квитків 1-го класу: {sum_for_1_class}')

sum_for_2_class = df[df['Pclass'] == 2]['Fare'].sum()
print(f'Сума цін квитків 2-го класу: {sum_for_2_class}')

sum_for_3_class = df[df['Pclass'] == 3]['Fare'].sum()
print(f'Сума цін квитків 3-го класу: {sum_for_3_class}')
```

```python
df['Family_Size'] = df['SibSp'] + df['Parch']

survived_by_family_size = df.groupby('Family_Size')['Survived'].mean()

print(survived_by_family_size)

plt.plot(survived_by_family_size.index, survived_by_family_size.values, marker='o', linestyle='-')
plt.xlabel("Розмір сім'ї")
plt.ylabel('Частота виживання')
plt.title("Частота виживання за розміром сім'ї")
plt.grid(True)
plt.xticks(survived_by_family_size.index)
plt.show()
```
